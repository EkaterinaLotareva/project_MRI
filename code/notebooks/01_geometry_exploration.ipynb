```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from src.geometry import points_on_rings_general, ring_center_general

print("✅ Все библиотеки загружены!")
```

```python
# Параметры
m = 6  # Количество стопок
n = 5  # Количество колец
radii = np.full(n, 0.035)  # Радиусы (м)
gaps = np.array([0.01, 0.001, 0.01, 0.001])  # Зазоры (м)
A = 0.08  # Расстояние до первого кольца (м)
N_seg = 200  # Точек на кольцо

print(f"Система: {m} стопок × {n} колец")
```

```python
# Расчёт координат
points = points_on_rings_general(
    delta=gaps, n=n, A=A, N=N_seg, R=radii, m=m
)

# 3D визуализация
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], s=5, alpha=0.6)
ax.set_xlabel('X (м)')
ax.set_ylabel('Y (м)')
ax.set_zlabel('Z (м)')
ax.set_title(f'Геометрия: {m} стопок × {n} колец')
plt.show()

print(f"Всего точек: {len(points)}")
```

```python
# 3D-визуализация

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:,0], points[:,1], points[:,2], s=1)
ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('z')
plt.show()
```
